```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
import pandas as pd
```

```python
## sets a linear through a regression manner
#label is used to label the line
#color sets the color of the line
#linewidth sets the width of the line
#marker sets the points
#markersize sets the point size
plt.plot([1,3,4], [2,4,6], label='random', color='gray', linewidth=3, marker='.', markersize=15)

## .title()

plt.title("first graph ever", fontdict={'fontname': 'sans', 'fontsize':20})
plt.xlabel("age")
plt.ylabel("pow")

## sets axis values for graph
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8])

## legend --> shows the label of the plot
plt.legend()

plt.show()

## REsize the graph, figsize = (X, Y) ratio
plt.figure(figsize=(5,3))

## saves plot as a png, you have to write name
plt.savefig()
```
